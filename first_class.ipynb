```python
# repeatedly prompt until we can open the named file
while True:
    name = input('Enter file name: ')
    try:
        handle = open(name, 'r')
        break
    except FileNotFoundError:
        print(f"No such file: {name!r}, please try again.")

count = dict()
for line in handle:
    words = line.split()      # fixed typo: split, not spilt
    for word in words:
        count[word] = count.get(word, 0) + 1

bigcount = None
bigword = None
for word, cnt in count.items():
    if bigcount is None or cnt > bigcount:
        bigword = word
        bigcount = cnt

print(bigword, bigcount)


```
